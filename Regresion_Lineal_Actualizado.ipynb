# **Modelos de Regresión**

Los modelos de regresión son una clase fundamental de métodos estadísticos y de aprendizaje automático que se utilizan para predecir una **variable objetivo (dependiente)** basándose en una o más **variables predictoras (independientes)**.

Estos modelos proporcionan una herramienta esencial para comprender y predecir comportamientos y tendencias basándose en datos históricos y observaciones.

Los orígenes de la regresión se remontan al siglo XIX, y desde entonces, estos modelos han encontrado aplicación en una amplia gama de campos.

### Orígenes y Desarrollo Histórico
1. **Francis Galton (1822-1911)**: El término "regresión" fue acuñado por el científico inglés Francis Galton en el contexto de sus estudios sobre herencia y características biológicas. Galton observó que, aunque las características físicas de los padres (como la altura) tendían a ser heredadas por sus hijos, también había una tendencia a "regresar" a la media de la población, un fenómeno que él llamó "regresión hacia la mediocridad" (más tarde conocido como "regresión a la media").

2. **Karl Pearson y Sir Francis Galton**: Karl Pearson, colaborador y sucesor de Galton, junto con otros estadísticos, desarrollaron formalmente los conceptos y métodos matemáticos de la regresión y la correlación.

### Tipos de Modelos de Regresión
- **Regresión Lineal Simple**: Usa una sola variable predictora para predecir una variable dependiente, con la relación modelada como una línea recta.
- **Regresión Lineal Múltiple**: Extiende la regresión lineal simple al incluir múltiples variables predictoras.
- **Regresión Logística**: Se usa para predecir una variable dependiente binaria (con dos posibles valores, como "sí" o "no").
- **Regresión Polinómica**: Modela la relación como un polinomio, permitiendo curvas en lugar de líneas rectas.
- **Otros tipos**: Incluyen regresión de Ridge, regresión LASSO, regresión no lineal, entre otros.

### Aplicaciones
Los modelos de regresión se utilizan en prácticamente todos los campos, desde la ciencia y la ingeniería hasta la economía y las ciencias sociales. Algunas aplicaciones comunes incluyen:

- **Predicción**: Estimar valores futuros basándose en datos históricos.
- **Análisis de tendencias**: Observar cómo cambian las cosas a lo largo del tiempo.
- **Control de calidad**: Evaluar la eficacia de los procesos de control de calidad.
- **Investigación científica**: Estudiar las relaciones entre variables en experimentos científicos.
- **Economía y negocios**: Predecir ventas, ingresos, demanda de productos, entre otros.

# **Regresión Lineal**

El **análisis de regresión**, especialmente la **regresión lineal**, se basa en varias suposiciones clave. Estas suposiciones son fundamentales para la validez de los resultados y conclusiones del análisis.

Estas suposiciones son cruciales para la interpretación adecuada y la validez de los resultados del análisis de regresión. Sin embargo, es importante recordar que en la práctica, es poco probable que todas estas suposiciones se cumplan perfectamente. En tales casos, se pueden aplicar técnicas de diagnóstico y métodos estadísticos alternativos para abordar las violaciones de estas suposiciones.

## **Suposiciones** principales:

1. **Relación Lineal**: La relación entre la variable dependiente y las variables independientes debe ser lineal. Esto significa que un cambio en una variable independiente se asocia con un cambio proporcional en la variable dependiente. Esta suposición se puede verificar visualmente usando diagramas de dispersión.

2. **Normalidad**: Se asume que los residuos (diferencias entre los valores observados y los valores predichos por el modelo) están distribuidos normalmente, especialmente en el contexto de las pruebas de hipótesis y la estimación de intervalos de confianza. La normalidad es menos crítica en grandes muestras debido al Teorema del Límite Central.

3. **No Multicolinealidad**: Las variables independientes no deben estar altamente correlacionadas entre sí. La multicolinealidad puede hacer que sea difícil aislar el efecto individual de una variable independiente y puede hacer que los coeficientes de la regresión sean inestables y difíciles de interpretar. Herramientas como el Factor de Inflación de la Varianza (VIF) pueden ayudar a detectar multicolinealidad.

4. **No Autocorrelación**: Los residuos no deben estar autocorrelacionados, especialmente importante en el análisis de series temporales. La autocorrelación puede ocurrir si los residuos en un período de tiempo están correlacionados con los residuos en otro período. La prueba de Durbin-Watson se utiliza comúnmente para detectar la autocorrelación.

5. **Varianzas Iguales (Homocedasticidad)**: Esta suposición implica que los residuos tienen varianzas constantes en diferentes niveles de las variables independientes. Si los residuos no tienen varianzas constantes (heterocedasticidad), puede afectar la eficiencia de los estimadores y hacer que las pruebas de hipótesis sean menos fiables. Las pruebas como la de Breusch-Pagan y los gráficos de residuos versus valores ajustados se utilizan para verificar la homocedasticidad.

## **Fórmula para calcular la Línea de Regresión**

La fórmula general para una línea de regresión lineal simple, que describe cómo una variable dependiente ($ y $) (por ejemplo, calificaciones) cambia con una variable independiente ($ x $) (por ejemplo, horas de estudio), se expresa como:

$ y = mx + b $

Donde:
- ($ y $) es la variable dependiente (la que queremos predecir).
- ($ x $) es la variable independiente (el predictor).
- ($ m $) es la pendiente de la línea de regresión, que indica cuánto cambia ($ y $) por cada unidad de cambio en ($ x $).
- ($ b $) es el intercepto con el eje y, que indica el valor de ($ y $) cuando ($ x $) es 0.

En el contexto de un modelo de regresión lineal, ($ m $) y ($ b $) se calculan a partir de los datos para minimizar la suma de los cuadrados de las diferencias entre los valores observados y los valores predichos por el modelo.

## **Calcular la Regresión Lineal**

La regresión lineal busca encontrar una ecuación de la forma $ y = mx + b $, donde "$ m $" es la pendiente de la línea y "$ b $" es la ordenada al origen. Para calcular estos valores, podemos usar fórmulas específicas:

La pendiente "m" se calcula utilizando la siguiente fórmula:

$ m = \frac{n(\sum(xy) - \sum x \sum y)}{n(\sum(x^2)) - (\sum x)^2} $

La ordenada al origen "b" se calcula de la siguiente manera:

$ b = \frac{\sum y - m(\sum x)}{n} $

Donde:
- $ n $ es el número de puntos de datos (en este caso, 5).
- $ ∑(xy) $ es la suma del producto de los valores de x e y.
- $ ∑x $ es la suma de los valores de x.
- $ ∑y $ es la suma de los valores de y.
- $ ∑(x^2) $ es la suma de los cuadrados de los valores de x.

Revisemos estos datos:

```
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

```

Primero, calculemos las sumas:

- $ ∑x = 1 + 2 + 3 + 4 + 5 = 15 $
- $ ∑y = 5 + 4 + 3 + 2 + 1 = 15 $
- $ ∑xy = (1*5) + (2*4) + (3*3) + (4*2) + (5*1) = 5 + 8 + 9 + 8 + 5 = 35 $
- $ ∑(x^2) = (1^2) + (2^2) + (3^2) + (4^2) + (5^2) = 1 + 4 + 9 + 16 + 25 = 55 $

Ahora, podemos calcular "m" y "b" utilizando las fórmulas mencionadas anteriormente:

$ m = \frac{5(35) - (15)(15)}{5(55) - (15)^2} = \frac{175 - 225}{275 - 225} = \frac{-50}{50} = -1 $

$ b = \frac{15 - (-1)(15)}{5} = \frac{15 + 15}{5} = \frac{30}{5} = 6 $

Entonces, la ecuación de regresión lineal que mejor se ajusta a tus datos es:

$ y = -1x + 6 $

Esta es la ecuación de la línea de regresión lineal que puedes usar para predecir valores de "y" para valores de "x" dentro del rango de tus datos.

## Computo y diagrama de dispersión de los datos

````python
import numpy as np

# Datos proporcionados
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 3, 2, 1])
n = len(x)

# Cálculo de la pendiente (m)
sum_xy = np.sum(x * y)
sum_x = np.sum(x)
sum_y = np.sum(y)
sum_x2 = np.sum(x**2)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)

# Cálculo de la ordenada al origen (b)
b = (sum_y - m * sum_x) / n

# Generar datos para la línea de regresión
x_line = np.linspace(min(x), max(x), 100)
y_line = m * x_line + b

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Datos reales')
plt.plot(x_line, y_line, color='red', label=f'Línea de regresión y={m:.2f}x+{b:.2f}')
plt.title('Gráfica de Regresión Lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

m, b  # Mostrar los valores de m y b
````

En este script se cargan las librerías necesarias para desarrollar el proyecto:

El siguiente script crea una serie temporal sintética de precios de combustible con componentes de ruido aleatorio, estacionalidad y tendencia.

1. **Configuración de la Semilla Aleatoria**:
   ```python
   np.random.seed(0)
   ```
   Esto establece la semilla para el generador de números aleatorios de NumPy. Usar una semilla específica asegura que los resultados sean reproducibles; cada vez que se ejecute el script con la misma semilla, generará la misma secuencia de números aleatorios.

2. **Definición de Parámetros de la Serie Temporal**:
   ```python
   periodos = 96  # 5 años de datos mensuales
   fecha_inicio = '2015-01'
   ```
   Aquí, se define la duración de la serie temporal. Se especifica que habrá 96 periodos (que en este caso son meses, por lo tanto, 8 años, no 5 como dice el comentario). La serie comienza en enero de 2015.

3. **Generación de Fechas**:
   ```python
   fechas = pd.date_range(start=fecha_inicio, periods=periodos, freq='M')
   ```
   Utilizando pandas, se crea un rango de fechas que comienza en `fecha_inicio` (`'2015-01'`) y se extiende por el número de `periodos` definido anteriormente. `freq='M'` indica que estas fechas son mensuales.

4. **Creación de Datos Sintéticos**:
   ```python
   precios_combustible = np.abs(np.random.normal(loc=3.0, scale=0.5, size=periodos) +
                                np.sin(np.linspace(-3, 3, periodos)) +
                                np.linspace(2.5, 3.5, periodos))
   ```
   Esta es la parte es donde se generan los datos sintéticos:
   - `np.random.normal(loc=3.0, scale=0.5, size=periodos)`: Genera datos aleatorios que siguen una distribución normal (o gaussiana) con una media (`loc`) de 3.0 y una desviación estándar (`scale`) de 0.5, con un total de `periodos` puntos de datos.
   - `np.sin(np.linspace(-3, 3, periodos))`: Crea un patrón senoidal. `np.linspace(-3, 3, periodos)` genera `periodos` puntos de datos equidistantes entre -3 y 3, y `np.sin()` aplica la función seno a estos puntos. Esto introduce una oscilación regular en los datos, simulando estacionalidad.
   - `np.linspace(2.5, 3.5, periodos)`: Genera una secuencia lineal de `periodos` puntos de datos equidistantes entre 2.5 y 3.5, añadiendo una tendencia lineal ascendente a los datos.
   - `np.abs(...)`: Asegura que todos los valores finales sean positivos, lo cual tiene sentido si estos representan precios.

````python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sintéticos
np.random.seed(0)
periodos = 96  # 5 años de datos mensuales
fecha_inicio = '2015-01'
fechas = pd.date_range(start=fecha_inicio, periods=periodos, freq='M')
precios_combustible = np.abs(np.random.normal(loc=3.0, scale=0.5, size=periodos) +
                             np.sin(np.linspace(-3, 3, periodos)) +
                             np.linspace(2.5, 3.5, periodos))

datos_sinteticos = pd.Series(precios_combustible, index=fechas)

# Descomposición estacional para visualizar los componentes
descomposicion = seasonal_decompose(datos_sinteticos, model='additive')
descomposicion.plot()
plt.show()

# Datos para visualizar
datos_sinteticos.head()
datos_sinteticos.to_csv('combustible.csv')
````

## Descomposición estacional

La descomposición estacional es una técnica utilizada en el análisis de series temporales para descomponer una serie temporal en varios componentes. Estos componentes típicamente incluyen:

**Tendencia**: Esta es la componente que muestra el patrón general o la dirección a largo plazo de los datos. Puede ser ascendente, descendente o constante.

**Estacionalidad**: Este componente refleja las variaciones o patrones que se repiten a intervalos regulares debido a factores estacionales. Por ejemplo, en el caso de los precios del combustible, podría haber un patrón estacional anual debido a la demanda variable en diferentes épocas del año.

**Residuo**: El residuo es lo que queda después de extraer la tendencia y la estacionalidad de la serie temporal. Esencialmente, representa la variabilidad aleatoria en los datos que no puede ser explicada por los componentes de tendencia y estacionalidad.

## **Entrenamiento del modelo de regresión para serie de tiempo utilizando el Modelo SARIMAX**

````python
# Dividir los datos en conjunto de entrenamiento y prueba
train = datos_sinteticos.iloc[:-12]  # Dejar los últimos 12 meses para la prueba
test = datos_sinteticos.iloc[-12:]

# Ajustar un modelo SARIMA
# Los parámetros del modelo se escogen de manera arbitraria para este ejemplo
modelo_sarima = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
resultado_sarima = modelo_sarima.fit()

# Predicciones
predicciones = resultado_sarima.get_forecast(steps=12)
predicciones_conf_int = predicciones.conf_int()
predicciones_media = predicciones.predicted_mean

# Graficar resultados
plt.figure(figsize=(10, 6))
plt.plot(train.index, train, label='Entrenamiento')
plt.plot(test.index, test, label='Verdadero', color='orange')
plt.plot(predicciones_media.index, predicciones_media, label='Predicciones', color='green')
plt.fill_between(predicciones_media.index,
                 predicciones_conf_int.iloc[:, 0],
                 predicciones_conf_int.iloc[:, 1], color='k', alpha=0.1)
plt.title('Predicción de Precios de Combustible con SARIMA')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.legend()
plt.show()

# Resumen del modelo
resultado_sarima.summary()
````

### **Métricas y estadísticas que aparecen en los resultados del modelo SARIMAX**

1. **Dep. Variable, No. Observations, Model, Date, Time, Sample, Covariance Type**: Estos son detalles generales del modelo, como la variable dependiente, el número de observaciones, el tipo de modelo (SARIMAX con los órdenes especificados), la fecha y hora del análisis, el período de la muestra de datos y el tipo de covarianza utilizado.

2. **Log Likelihood**: Es una medida de cuán bien el modelo se ajusta a los datos. Un valor más alto indica un mejor ajuste.

3. **AIC (Criterio de Información de Akaike)**: Ayuda a comparar modelos; un valor más bajo indica un mejor modelo. Es útil para la selección de modelos, especialmente cuando se comparan modelos con diferentes números de parámetros.

4. **BIC (Criterio de Información Bayesiano)**: Al igual que el AIC, se utiliza para la selección de modelos. Penaliza modelos con más parámetros más severamente que el AIC.

5. **HQIC (Criterio de Información de Hannan-Quinn)**: Otra métrica para comparar modelos, similar al AIC y BIC.

6. **Coeficientes (coef) y Error Estándar (std err)**: Los coeficientes son los valores estimados para cada término del modelo. El error estándar mide la variabilidad o incertidumbre de estas estimaciones.

7. **Z y P>|z|**: La estadística 'z' y el valor 'P' asociado son pruebas de significancia de los coeficientes. Un valor 'P' bajo indica que el coeficiente es significativamente diferente de cero.

8. **Intervalo de Confianza [0.025 0.975]**: Este intervalo indica dónde se espera que se encuentre el verdadero valor del coeficiente con un 95% de confianza.

9. **Ljung-Box (Q), Prob(Q)**: Prueba la autocorrelación de los residuos. Un valor alto de Prob(Q) sugiere que los residuos no están autocorrelacionados, lo cual es bueno.

10. **Jarque-Bera (JB), Prob(JB)**: Prueba la normalidad de los residuos. Un valor alto de Prob(JB) sugiere que los residuos están normalmente distribuidos.

11. **Heteroskedasticity (H), Prob(H)**: Prueba la homogeneidad de la varianza de los residuos. Un valor alto en Prob(H) sugiere que no hay heteroscedasticidad.

12. **Skew y Kurtosis**: Medidas de la asimetría y la forma de la distribución de los residuos, respectivamente.

En el modelo, los valores altos en los p-valores de los coeficientes indican que muchos de ellos no son estadísticamente significativos, lo que podría señalar la necesidad de ajustar el modelo. Además, las métricas de diagnóstico, como Ljung-Box, indican que los residuos no presentan problemas evidentes de autocorrelación, lo cual es positivo.
